```python
from sklearn import manifold
from skimage.io import imread_collection
import cv2
import numpy as np

DATA_DIR = "preproc2/*"
img_data = imread_collection(DATA_DIR, cv2.IMREAD_GRAYSCALE)
```

```python
import glob
names = glob.glob("preproc2/*.png")
```

```python
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np

imgs = []
for name in names:
    imgs.append(image.load_img(name, target_size=(128, 128)))

arrays_of_images = []
flattened = []
for img in imgs:
    arrays_of_images.append(image.img_to_array(img))
    flattened.append(image.img_to_array(img).flatten())
#tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)
#X_tsne = tsne.fit_transform(vgg16_feature)
```

```python
flattened = np.asarray(flattened)
flattened.shape
```

```python
arrays_of_images = np.asarray(arrays_of_images)
model = VGG16(weights='imagenet', include_top=False)
vgg16_feature = model.predict(arrays_of_images)
```

```python
tsne = manifold.TSNE(n_components=2, learning_rate=20, n_iter=5000, init='pca' random_state=0)
X_tsne = tsne.fit_transform(flattened)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import offsetbox

def plot_embedding(X, title=None):
    x_min, x_max = np.min(X, 0), np.max(X, 0)
    X = (X - x_min) / (x_max - x_min)

    plt.figure(figsize=(32, 32))
    ax = plt.subplot(111)
    for i in range(X.shape[0]):
        plt.text(X[i, 0], X[i, 1], names[i].split('_')[2].split('.')[0],
                 color=plt.cm.Set1(int(names[i].split('_')[2].split('.')[0]) / 10.),
                 fontdict={'weight': 'bold', 'size': 9})

    if hasattr(offsetbox, 'AnnotationBbox'):
        # only print thumbnails with matplotlib > 1.0
        shown_images = np.array([[1., 1.]])  # just something big
        for i in range(X.shape[0]):
            dist = np.sum((X[i] - shown_images) ** 2, 1)
            if np.min(dist) < 4e-3:
                # don't show points that are too close
                continue
            shown_images = np.r_[shown_images, [X[i]]]
            imagebox = offsetbox.AnnotationBbox(
                offsetbox.OffsetImage(imgs[i], cmap=plt.cm.gray_r),
                X[i])
            #ax.add_artist(imagebox)
    plt.xticks([]), plt.yticks([])
    if title is not None:
        plt.title(title)
```

```python
plot_embedding(X_tsne,
               "t-SNE embedding(time %.2fs)" %
               (1))
```
