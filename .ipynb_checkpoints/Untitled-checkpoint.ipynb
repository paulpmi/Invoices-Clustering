```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from skimage.measure import find_contours, approximate_polygon, \
    subdivide_polygon

import cv2
DATA_DIR = 'data/'
```

```python
im = cv2.imread("table_images/table_467.jpg", cv2.IMREAD_GRAYSCALE)
im.shape
fig, ax = plt.subplots()
ax.imshow(im, cmap=plt.cm.gray)
```

```python
cv2.imshow('ImageWindow', im)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
detector = cv2.SimpleBlobDetector_create()
keypoints = detector.detect(im)
im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (0,0,255), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2.imshow("Keypoints", im_with_keypoints)
cv2.waitKey(0)
```

```python
contours = find_contours(im, 0.8)

# Display the image and plot all contours found

for n, contour in enumerate(contours):
    ax.plot(contour[:, 1], contour[:, 0], linewidth=2)

ax.axis('image')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
for contour in find_contours(im, 0):
    coords = approximate_polygon(contour, tolerance=2.5)
    ax.plot(coords[:, 1], coords[:, 0], '-r', linewidth=2)
    coords2 = approximate_polygon(contour, tolerance=39.5)
    ax.plot(coords2[:, 1], coords2[:, 0], '-g', linewidth=2)
    print("Number of coordinates:", len(contour), len(coords), len(coords2))

ax.axis((0, 800, 0, 800))

plt.show()
```

```python
from skimage.filters import hessian

result = hessian(im, scale_step=10, alpha=2, beta=10, gamma=20, black_ridges=False)
plt.imshow(result)
```

```python
from skimage import feature
edges1 = feature.canny(result, sigma=1.5)
plt.imshow(edges1)
```

```python
from skimage.filters import roberts, sobel, gaussian

#edgesr = roberts(im)
edgess = sobel(im)
gauss = gaussian(im,  sigma=1)
gauss_sobel = sobel(gauss)
#plt.imshow(edgesr)
plt.imshow(edgess)
```

```python
from skimage.segmentation import slic
from skimage.future import graph
from skimage import io, color

labels = slic(im, compactness=30, n_segments=400)

g = graph.rag_boundary(labels, edgess)
edges_rgb = color.gray2rgb(edgess)
lc = graph.show_rag(labels, g, edges_rgb, img_cmap=None, edge_cmap='viridis',
                    edge_width=1.2)

plt.colorbar(lc, fraction=0.03)
io.show()
```

```python
from skimage.restoration import denoise_bilateral, denoise_wavelet, estimate_sigma


sigma_est = estimate_sigma(gauss_sobel, multichannel=False, average_sigmas=True)
sigma_est
```

```python
denoised = denoise_wavelet(gauss_sobel, multichannel=False)

plt.imshow(denoised)
```

```python
from skimage.morphology import watershed
from scipy import ndimage as ndi
distance = ndi.distance_transform_edt(im)
from skimage.feature import peak_local_max
local_maxi = peak_local_max(distance, labels=im,
                             footprint=np.ones((3, 3)),
                             indices=False)
markers = ndi.label(local_maxi)[0]
segmentation = watershed(gauss_sobel, markers)
plt.imshow(segmentation)
```
